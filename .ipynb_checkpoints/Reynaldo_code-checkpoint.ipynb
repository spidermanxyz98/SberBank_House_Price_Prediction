```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn import model_selection, preprocessing
import xgboost as xgb
import datetime
import operator

from sklearn import model_selection, preprocessing
import xgboost as xgb
import datetime
#now = datetime.datetime.now()

train = pd.read_csv('input/train.csv')
test = pd.read_csv('input/test.csv')
macro = pd.read_csv('input/macro.csv')
id_test = test.id
train.sample(3)
```

```python
y_train = train["price_doc"]
x_train = train.drop(["id", "timestamp", "price_doc"], axis=1)
x_test = test.drop(["id", "timestamp"], axis=1)

for c in x_train.columns:
    if x_train[c].dtype == 'object':
        lbl = preprocessing.LabelEncoder()
        lbl.fit(list(x_train[c].values)) 
        x_train[c] = lbl.transform(list(x_train[c].values))
        #x_train.drop(c,axis=1,inplace=True)
        
for c in x_test.columns:
    if x_test[c].dtype == 'object':
        lbl = preprocessing.LabelEncoder()
        lbl.fit(list(x_test[c].values)) 
        x_test[c] = lbl.transform(list(x_test[c].values))
        #x_test.drop(c,axis=1,inplace=True)        
```

```python
# xgb_params = {
#     'eta': 0.05,
#     'max_depth': 5,
#     'subsample': 0.7,
#     'colsample_bytree': 0.7,
#     'objective': 'reg:linear',
#     'eval_metric': 'rmse',
#     'silent': 1
# }

#try different params
xgb_params = {
    'eta': 0.02,
    'max_depth': 6,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'objective': 'reg:linear',
    'eval_metric': 'rmse',
    'silent': 1,
    'labmda':100
}

dtrain = xgb.DMatrix(x_train, y_train)
dtest = xgb.DMatrix(x_test)
```

```python
cv_output = xgb.cv(xgb_params, dtrain, num_boost_round=1000, early_stopping_rounds=20,
    verbose_eval=50, show_stdv=True, nfold=5)
#cv_output[['train-rmse-mean', 'test-rmse-mean']].plot()
```

```python
num_boost_rounds = len(cv_output)
model = xgb.train(dict(xgb_params, silent=0), dtrain, num_boost_round= num_boost_rounds)
```

```python
# fig, ax = plt.subplots(1, 1, figsize=(8, 46))
# xgb.plot_importance(model, height=0.5, ax=ax)
# plt.savefig('plots/Reynaldo_fscore.pdf')
```

```python
y_predict = model.predict(dtest)
output = pd.DataFrame({'id': id_test, 'price_doc': y_predict})
output.head()
```

```python
output.to_csv('170530_Reynaldo_code.csv', index=False)
```

# Tune XGB model

```python
# score=model.get_fscore()
# score=sorted(score.items(), key=operator.itemgetter(1), reverse=True)
# score=pd.DataFrame(score, columns=['feature', 'fscore'])
# score.to_csv(path_or_buf='submissions/scores/Reynaldo_xgb_score.csv')
score=pd.read_csv('submissions/scores/Reynaldo_xgb_score.csv')


importance_rate=0.01
f_index=score[score.fscore>importance_rate*score.fscore.max()].feature.values
print('number of f_index', len(f_index))
```

```python
#train with important features

Dtrain_select=xgb.DMatrix(x_train[f_index], y_train)
Dtest_select=xgb.DMatrix(x_test[f_index])

new_xgb_params={
    'eta': 0.02,
    'max_depth': 6,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'objective': 'reg:linear',
    'eval_metric': 'rmse',
    'silent': 1,
    'labmda':1
}
new_model_cv=xgb.cv(new_xgb_params, Dtrain_select, nfold=5,
                   verbose_eval=100, num_boost_round=2000, early_stopping_rounds=20)
```

```python
num_boost_round=len(new_model_cv)
new_model=xgb.train(new_xgb_params, Dtrain_select, num_boost_round=num_boost_round)
```

```python
y_predict = new_model.predict(Dtest_select)
output = pd.DataFrame({'id': id_test, 'price_doc': y_predict})
output.head()
```

```python
output.to_csv('170530_Reynalde_0.01_xgb_importance.csv', index=False)
```
